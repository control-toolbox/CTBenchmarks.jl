*The Robot Arm Problem:* We aim to Minimize the time taken for a robot arm to travel between two points.                
The problem is inspired from the following paper: [Benchmarking Optimization Software with COPS 3.0](https://www.mcs.anl.gov/~more/cops/cops3.pdf).                             
We solve the problem using the [JuMP](https://jump.dev/) and [OptimalControl](https://control-toolbox.org/OptimalControl.jl/stable/) packages in Julia then we compare the results.

1. [JuMP Formulation](#jump)
2. [OptimalControl Formulation](#oc)
3. [Comparison](#comp)

</div>

```julia
import Pkg
Pkg.activate("..")
include("./utils.jl");
using OptimalControlProblems
using Plots
using MKL
import HSL_jll
nh = 100;
```

## 1. Solving with JuMP <a id='jump'></a>

```julia
using JuMP, Ipopt

RobotJMP = robot(JuMPBackend();nh=nh)

set_optimizer(RobotJMP,Ipopt.Optimizer)
set_attribute(RobotJMP, "print_level", 5)
set_optimizer_attribute(RobotJMP,"tol",1e-8)
set_optimizer_attribute(RobotJMP,"constr_viol_tol",1e-6)
set_optimizer_attribute(RobotJMP,"max_iter",1000)
set_optimizer_attribute(RobotJMP,"mu_strategy","adaptive")
set_attribute(RobotJMP, "hsllib", HSL_jll.libhsl_path)
set_attribute(RobotJMP, "linear_solver", "ma57")
set_optimizer_attribute(RobotJMP, "print_timing_statistics", "yes")
```

```julia
println("Solving...")
optimize!(RobotJMP)
println()

# Display results
if termination_status(RobotJMP) == MOI.OPTIMAL
    println("  Solution is optimal")
elseif  termination_status(RobotJMP) == MOI.LOCALLY_SOLVED
    println("  (Local) solution found")
elseif termination_status(RobotJMP) == MOI.TIME_LIMIT && has_values(RobotJMP)
    println("  Solution is suboptimal due to a time limit, but a primal solution is available")
else
    error("  The model was not solved correctly.")
end
println("  objective value = ", objective_value(RobotJMP))
println()

rho_ = RobotJMP[:rho];
the_ = RobotJMP[:the];
phi_ = RobotJMP[:phi];
rho_dot_ = RobotJMP[:rho_dot];
the_dot_ = RobotJMP[:the_dot];
phi_dot_ = RobotJMP[:phi_dot];
u_rho_ = RobotJMP[:u_rho];
u_the_ = RobotJMP[:u_the];
u_phi_ = RobotJMP[:u_phi];
con_rho = RobotJMP[:con_rho];
con_the = RobotJMP[:con_the];
con_phi = RobotJMP[:con_phi];
con_rho_dot = RobotJMP[:con_rho_dot];
con_the_dot = RobotJMP[:con_the_dot];
con_phi_dot = RobotJMP[:con_phi_dot];

tf = value.(RobotJMP[:tf])
t = Vector((0:nh)*tf/nh);

p = -[[ dual(con_rho[i]), dual(con_the[i]), dual(con_phi[i]), dual(con_rho_dot[i]), dual(con_the_dot[i]), dual(con_phi_dot[i]) ] for i in 2:nh+1];
p = costateInterpolation(p, t);

rho = Vector(value.(rho_)) ;
the = Vector(value.(the_)) ;
phi = Vector(value.(phi_)) ;
phi_dot = Vector(value.(phi_dot_)) ;
the_dot = Vector(value.(the_dot_)) ;
rho_dot = Vector(value.(rho_dot_)) ;

u_phi = Vector(value.(u_phi_)) ;
u_the = Vector(value.(u_the_)) ;
u_rho = Vector(value.(u_rho_)) ;

u = [ [ u_rho[i], u_the[i], u_phi[i] ] for i in 1:nh+1 ];
x =  [ [ rho[i],rho_dot[i],the[i],the_dot[i],phi[i],phi_dot[i] ] for i in 1:nh+1 ];
```

```julia
x1_plot = plot(t, [ x[i][1] for i in 1:nh+1 ], xlabel = "t", ylabel = "rho", legend = false)
x2_plot = plot(t, [ x[i][2] for i in 1:nh+1 ], xlabel = "t", ylabel = "drho", legend = false)
x3_plot = plot(t, [ x[i][3] for i in 1:nh+1 ], xlabel = "t", ylabel = "the", legend = false)
x4_plot = plot(t, [ x[i][4] for i in 1:nh+1 ], xlabel = "t", ylabel = "dthe", legend = false)
x5_plot = plot(t, [ x[i][5] for i in 1:nh+1 ], xlabel = "t", ylabel = "phi", legend = false)
x6_plot = plot(t, [ x[i][6] for i in 1:nh+1 ], xlabel = "t", ylabel = "dphi", legend = false)
px1_plot = plot(t, [ p[i][1] for i in 1:nh+1 ], xlabel = "t", ylabel = "px1", legend = false)
px2_plot = plot(t, [ p[i][2] for i in 1:nh+1 ], xlabel = "t", ylabel = "px2", legend = false)
px3_plot = plot(t, [ p[i][3] for i in 1:nh+1 ], xlabel = "t", ylabel = "px3", legend = false)
px4_plot = plot(t, [ p[i][4] for i in 1:nh+1 ], xlabel = "t", ylabel = "px4", legend = false)
px5_plot = plot(t, [ p[i][5] for i in 1:nh+1 ], xlabel = "t", ylabel = "px5", legend = false)
px6_plot = plot(t, [ p[i][6] for i in 1:nh+1 ], xlabel = "t", ylabel = "px6", legend = false)

u1_plot = plot(t, [ u[i][1] for i in 1:nh+1 ], xlabel = "t", ylabel = "u", legend = false)
u2_plot = plot(t, [ u[i][2] for i in 1:nh+1 ], xlabel = "t", ylabel = "u", legend = false)
u3_plot = plot(t, [ u[i][3] for i in 1:nh+1 ], xlabel = "t", ylabel = "u", legend = false)

layout = @layout [a b; c d; e f; g h; i j; k l; m; o; p]
x_plot = plot(x1_plot, px1_plot, x2_plot, px2_plot, x3_plot, px3_plot, 
            x4_plot, px4_plot, x5_plot, px5_plot, x6_plot, px6_plot,
            u1_plot, u2_plot, u3_plot, layout = layout)
x_plot = plot!(size = (800, 1200))

display(x_plot)
```

## 2. Solving with OptimalControl <a id='oc'></a>

```julia
using OptimalControl, NLPModelsIpopt
docp, RobotOC = robot(OptimalControlBackend();nh=nh);
```

```julia
# Solve the problem
nlp_sol = NLPModelsIpopt.ipopt(
    RobotOC;
    print_level=5,
    tol=1e-8,
    mu_strategy="adaptive",
    sb="yes",
    constr_viol_tol=1e-6,
    max_iter=1000,
    print_timing_statistics="yes",
    linear_solver="ma57",
    hsllib=HSL_jll.libhsl_path,
)
```

```julia
# Plot results
sol = OptimalControlSolution(docp; primal=nlp_sol.solution, dual=nlp_sol.multipliers)
plot(sol)
```
